## Clustering

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = ["Sepal_length", "Sepal_width", "Petal_length", "Petal_width"])
df
```

```python
iris.target_names
iris.feature_names
```

```python
features = ["Petal_length", "Sepal_length", "Petal_width", "Sepal_width"] 
```

```python
from sklearn.cluster import KMeans 

kmeans = KMeans(n_clusters = 3, random_state = 42)
kmeans
```

```python
kmeans.fit(df[features])
```

```python
y_pred = kmeans.predict(df[features])
y_pred
```

```python
df["Cluster"] = y_pred
df
```

```python
df1 = df[df.Cluster == 0]
df2 = df[df.Cluster == 1]
df3 = df[df.Cluster == 2]
df1 ,df2, df3 
```

```python
plt.scatter(df1.Petal_length, df1.Sepal_length, color='red')
plt.scatter(df2.Petal_length, df2.Sepal_length, color='black')
plt.scatter(df3.Petal_length, df3.Sepal_length, color='green')

plt.xlabel('Petal length')
plt.ylabel('Sepal_length')
plt.legend()
plt.show()
```

```python
plt.scatter(df1.Petal_length, df1.Sepal_length, color='red')
plt.scatter(df2.Petal_length, df2.Sepal_length, color='black')
plt.scatter(df3.Petal_length, df3.Sepal_length, color='green')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='blue', label='centroid') 

plt.xlabel('Petal length')
plt.ylabel('Sepal_length')
plt.legend()

plt.show()
```

### Elbow method

```python
WCSS = []
k_range = range(1, 10)

features = ["Petal_length", "Sepal_length", "Petal_width", "Sepal_width"] 
for k in k_range: 
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(df[features])
    WCSS.append(kmeans.inertia_) 
WCSS
```

WCSS represents the sum of squared distances between each point and the centroid of its assigned cluster. 'wcss' will contain the WCSS values for each k from 1 to 9.

```python
plt.plot(k_range, WCSS)
plt.xlabel("k")
plt.ylabel("WCSS") 
plt.title("Elbow Plot")
plt.show()
```

From this plot, it is clear that k = 3, as the elbow shape forms at k = 3 after which the values decrease gradually.

### Silhouette Coefficient

A higher silhouette score indicates better clustering.

```python
from sklearn.metrics import silhouette_score

S = []
k_rnge = range(2, 11)

features = ["Petal_length", "Sepal_length", "Petal_width", "Sepal_width"] 
for k in k_rnge: 
    kmeans = KMeans(n_clusters = k)
    km_pred = kmeans.fit_predict(df[features])
    S.append(silhouette_score(df[features], km_pred)) 
S
```

```python
plt.plot(k_range, S)
plt.xlabel("Clusters")
plt.ylabel("Silhouette score") 
plt.title("Silhouette Score v/s Cluster")
plt.show()
```

Here, when cluster k = 2, the Silhouette score appears to be the highest. So the number of clusters needed is 2.
